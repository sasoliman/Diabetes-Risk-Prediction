# Diabetes Risk Prediction

## Notebook 01 – Data Loading

### Objectives

This notebook downloads the CDC Diabetes Health Indicators dataset from the UCI Machine Learning Repository, performs an initial inspection, and stores the raw dataset for downstream preprocessing, exploratory data analysis, and machine learning.

### Outputs

- data/raw/diabetes_raw.parquet
- results/dataset_metadata.txt
- results/variable_information.csv

```python
# ===============================================================
# Import Libraries
# ===============================================================

from pathlib import Path

import pandas as pd

from ucimlrepo import fetch_ucirepo

import pyarrow
```

```python
# ===============================================================
# Project Configuration
# ===============================================================

from pathlib import Path

def find_project_root(start: Path | None = None) -> Path:
    """Locate the project root independent of the Jupyter kernel's working
    directory.

    In VS Code, a notebook's working directory depends on the
    `jupyter.notebookFileRoot` setting (commonly either the notebook's own
    folder, or the workspace root) and this can silently differ between
    notebooks or sessions. Assuming `PROJECT_ROOT = Path.cwd().parent` breaks
    the moment two notebooks in the same project are launched under
    different conventions -- one notebook writes files under one resolved
    root, the other reads from a different one, and you get a confusing
    FileNotFoundError that looks like a missing-data problem but is really
    a working-directory mismatch.

    This walks upward from `start` (default: cwd) looking for either:
      - a directory literally named 'notebooks' -> its parent is the root, or
      - a directory that itself contains a 'notebooks' subfolder -> that IS the root.
    Both common cwd conventions resolve to the same PROJECT_ROOT this way.
    """
    current = (start or Path.cwd()).resolve()
    for candidate in (current, *current.parents):
        if candidate.name == "notebooks":
            return candidate.parent
        if (candidate / "notebooks").is_dir():
            return candidate
    raise FileNotFoundError(
        f"Could not locate the project root by walking up from {current}. "
        "Expected to find a 'notebooks' directory at some level above or "
        "at the current working directory."
    )

PROJECT_ROOT = find_project_root()
print(f"Resolved PROJECT_ROOT: {PROJECT_ROOT}")

DATA_DIR = PROJECT_ROOT / "data"

RAW_DIR = DATA_DIR / "raw"

RESULTS_DIR = PROJECT_ROOT / "results"

RAW_DIR.mkdir(
    parents=True,
    exist_ok=True
)

RESULTS_DIR.mkdir(
    parents=True,
    exist_ok=True
)
```

```python
# ===============================================================
# Download Dataset
# ===============================================================

cdc_diabetes_health_indicators = fetch_ucirepo(id=891)

X = cdc_diabetes_health_indicators.data.features

y = cdc_diabetes_health_indicators.data.targets

print("Dataset downloaded successfully.")
```

```python
# ===============================================================
# Create DataFrame
# ===============================================================

df = pd.concat(

    [X, y],

    axis=1

)

```

```python
# ===============================================================
# Dataset Preview
# ===============================================================

print(df.shape)

display(df.head())

display(df.info())
```

```python
# ===============================================================
# Save Raw Dataset
# ===============================================================

raw_data_path = RAW_DIR / "diabetes_raw.parquet"

df.to_parquet(
    raw_data_path,
    index=False
)

print("Raw dataset saved successfully.")

print(f"Location : {raw_data_path}")
```

```python
# ===============================================================
# Save Dataset Metadata
# ===============================================================

metadata_path = RESULTS_DIR / "dataset_metadata.txt"

with open(metadata_path, "w", encoding="utf-8") as file:

    file.write(str(cdc_diabetes_health_indicators.metadata))

print("Metadata saved successfully.")

print(f"Location : {metadata_path}")
```

```python
# ===============================================================
# Save Variable Information
# ===============================================================

variable_information = cdc_diabetes_health_indicators.variables

variable_path = RESULTS_DIR / "variable_information.csv"

variable_information.to_csv(
    variable_path,
    index=False
)

print("Variable information saved successfully.")

print(f"Location : {variable_path}")
```

```python
# ===============================================================
# Verify Saved Files
# ===============================================================

print("=" * 70)
print("Project Outputs")
print("=" * 70)

print(f"Raw Dataset         : {raw_data_path}")
print(f"Dataset Metadata    : {metadata_path}")
print(f"Variable Information: {variable_path}")
```
