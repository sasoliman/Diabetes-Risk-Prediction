```python
# ===============================================================
# Import Libraries
# ===============================================================

from pathlib import Path

import pandas as pd

from ucimlrepo import fetch_ucirepo
```

```python
# ===============================================================
# Project Configuration
# ===============================================================

# Project Root
PROJECT_ROOT = Path.cwd().parent

# Data
DATA_DIR = PROJECT_ROOT / "data"
RAW_DATA_DIR = DATA_DIR / "raw"

# Results
RESULTS_DIR = PROJECT_ROOT / "results"

# Create folders if they do not exist
RAW_DATA_DIR.mkdir(parents=True, exist_ok=True)
RESULTS_DIR.mkdir(parents=True, exist_ok=True)
```

```python
# ===============================================================
# Download Dataset
# ===============================================================

cdc_diabetes = fetch_ucirepo(id=891)

X = cdc_diabetes.data.features

y = cdc_diabetes.data.targets

df = pd.concat([X, y], axis=1)

print(f"Dataset Shape: {df.shape}")
```

```python
# ===============================================================
# Dataset Preview
# ===============================================================

display(df.head())

display(df.info())
```

```python
# ===============================================================
# Save Raw Dataset
# ===============================================================

raw_data_path = RAW_DATA_DIR / "diabetes_raw.csv"

df.to_csv(
    raw_data_path,
    index=False
)

print(f"Raw dataset saved to:\n{raw_data_path}")
```

```python
# ===============================================================
# Save Dataset Metadata
# ===============================================================

metadata_path = RESULTS_DIR / "dataset_metadata.txt"

with open(metadata_path, "w", encoding="utf-8") as file:

    file.write(str(cdc_diabetes.metadata))

print(f"Metadata saved to:\n{metadata_path}")
```

```python
# ===============================================================
# Save Variable Information
# ===============================================================

variable_information = cdc_diabetes.variables

variable_path = RESULTS_DIR / "variable_information.csv"

variable_information.to_csv(
    variable_path,
    index=False
)

print(f"Variable information saved to:\n{variable_path}")
```
