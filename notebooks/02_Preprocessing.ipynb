# Diabetes Risk Prediction

## 02 - Data Preprocessing

### Objectives

This notebook prepares the raw dataset for downstream statistical analysis and machine learning by:

- Converting variables to appropriate data types
- Creating clinically meaningful categorical variables
- Assessing data quality
- Saving the processed dataset

### Input

data/raw/diabetes_raw.csv

### Output

data/processed/diabetes_processed.csv

```python
# ===============================================================
# Import Libraries
# ===============================================================

from pathlib import Path

import numpy as np
import pandas as pd
```

```python
# ===============================================================
# Project Configuration
# ===============================================================

PROJECT_ROOT = Path.cwd().parent

DATA_DIR = PROJECT_ROOT / "data"

RAW_DATA_DIR = DATA_DIR / "raw"

PROCESSED_DATA_DIR = DATA_DIR / "processed"

RESULTS_DIR = PROJECT_ROOT / "results"

PROCESSED_DATA_DIR.mkdir(
    parents=True,
    exist_ok=True
)

RESULTS_DIR.mkdir(
    parents=True,
    exist_ok=True
)
```

```python
# ===============================================================
# Load Raw Dataset
# ===============================================================

raw_data_path = RAW_DATA_DIR / "diabetes_raw.csv"

df = pd.read_csv(raw_data_path)

print(f"Dataset Shape: {df.shape}")

display(df.head())
```

```python
# ===============================================================
# Initial Data Inspection
# ===============================================================

print("=" * 80)
print("Dataset Information")
print("=" * 80)

df.info()

print("\n")

print("=" * 80)
print("Summary Statistics")
print("=" * 80)

display(df.describe(include="all"))
```

```python
# ===============================================================
# Convert Data Types
# ===============================================================

categorical_columns = [

    "Diabetes_binary",

    "HighBP",

    "HighChol",

    "Smoker",

    "Stroke",

    "HeartDiseaseorAttack",

    "PhysActivity",

    "Fruits",

    "Veggies",

    "HvyAlcoholConsump",

    "AnyHealthcare",

    "NoDocbcCost",

    "GenHlth",

    "MentHlth",

    "PhysHlth",

    "DiffWalk",

    "Sex",

    "Education",

    "Income",

    "CholCheck",

    "Age"

]

df[categorical_columns] = (

    df[categorical_columns]

    .astype("category")

)

print(df.dtypes)
```

```python
# ===============================================================
# Create BMI Categories
# ===============================================================

bmi_bins = [

    -np.inf,

    18.5,

    25,

    30,

    np.inf

]

bmi_labels = [

    "Underweight",

    "Normal weight",

    "Overweight",

    "Obese"

]

df["BMI_cat"] = pd.cut(

    df["BMI"],

    bins=bmi_bins,

    labels=bmi_labels,

    ordered=True

)
```

```python
# ===============================================================
# Create Mental Health Categories
# ===============================================================

df["MentHlth_cat"] = pd.cut(

    df["MentHlth"],

    bins=[-1, 0, 10, 20, 30],

    labels=[

        "0 days",

        "1-10 days",

        "11-20 days",

        "21-30 days"

    ]

)
```

```python
# ===============================================================
# Create Physical Health Categories
# ===============================================================

df["PhysHlth_cat"] = pd.cut(

    df["PhysHlth"],

    bins=[-1, 0, 10, 20, 30],

    labels=[

        "0 days",

        "1-10 days",

        "11-20 days",

        "21-30 days"

    ]

)
```

```python
# ===============================================================
# Data Quality Assessment
# ===============================================================

print("=" * 80)
print("Data Quality Assessment")
print("=" * 80)

print(f"Observations : {df.shape[0]}")

print(f"Variables    : {df.shape[1]}")

print(f"Duplicates   : {df.duplicated().sum()}")

print("\n")

print("Missing Values")

display(df.isnull().sum())

print("\n")

print("Data Types")

display(df.dtypes)
```

```python
# ===============================================================
# Save Processed Dataset
# ===============================================================

processed_path = (

    PROCESSED_DATA_DIR

    / "diabetes_processed.csv"

)

df.to_csv(

    processed_path,

    index=False

)

print(f"Processed dataset saved to:\n{processed_path}")
```
